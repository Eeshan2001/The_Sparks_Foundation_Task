```python
#import the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# load the iris datasets
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
y = iris.target
iris_df.head(10)
```

```python
iris_df.info()
```

```python
iris_df.describe()
```

## define the decision tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(iris_df,y)

print('Decision Tree Classifer Created')
```

```python
from sklearn import tree
import matplotlib.pyplot as plt


featName=['sepal length ( in cm)','sepal width (in cm)','petal length (in cm)','petal width (in cm)']
clsName=['Iris-Setosa','Iris-Versicolor','Iris-virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,6), dpi = 350)

tree.plot_tree(dtree, feature_names = featName, class_names = clsName, filled = True);
```
