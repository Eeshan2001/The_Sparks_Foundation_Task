```python
#Importing required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
df = pd.read_csv( "SampleSuperstore.csv")
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## Cleaning of dataset and removing the duplicate data

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## Exploratory Data Analysis

```python
# Using corelation Matrix
plt.figure(figsize=(7,6))
correlation=df.corr()
sns.heatmap(correlation,annot=True,cmap='Reds')
```

## Line Plot

```python
plt.figure(figsize=(15,5))
plt.plot(df['Profit'], color = 'blue')
plt.grid(alpha=0.2)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=df, color = 'Purple')
plt.legend('Profit')
```

```python
df.corr()
```

## Distribution Plot

```python

fig, axs = plt.subplots(ncols=2, nrows=2,figsize=(10,10))
sns.distplot(df['Sales'], color = 'Green', ax=axs[0][0])
sns.distplot(df['Profit'], color = 'Red', ax=axs[0][1])
sns.distplot(df['Quantity'], color = 'Blue', ax=axs[1][0])
sns.distplot(df['Discount'], color = 'Orange', ax=axs[1][1])


axs[0][0].set_title('Sales Distribution', fontsize = 20  )
axs[0][1].set_title('Profit Distribution', fontsize = 20 )
axs[1][0].set_title('Quantity Distribution', fontsize = 20 )
axs[1][1].set_title('Discount Distribution', fontsize = 20 )

plt.show()
```

## State wise analysis of Profit,Discount,Sale 

```python
df['State'].value_counts().head(10)
```

```python
df_state = df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

```python
df_state1=df_state.sort_values('Profit')
```

```python
df_state1[['Profit']].plot(kind='bar',figsize=(20,8), color='blue')
plt.title('State wise Profit Analysis', fontsize = 22)
plt.ylabel('Profit per State', fontsize = 20)
plt.xlabel('States', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct= '%3.1f%%')
plt.show()
```

## The top subcategories are Binders, Papers ,Furnishings, Phones, Storage , Art and Accessories while the bottom subcategories are copiers and machines.

## Using Cat Plot

```python
## Observing the top selling states
top_15 = df.groupby('State')['Quantity'].sum().reset_index().sort_values(by = 'Quantity', ascending = False)
sns.catplot(x = 'State', y = 'Quantity', data = top_15.head(15), kind = 'bar', palette = 'inferno', aspect = 3)
plt.title('Top 15 States', fontsize = 22)
plt.xticks(rotation = 50, ha = 'right', fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('States', fontsize = 20)
plt.ylabel('Amount ordered', fontsize = 20)
```

## We observe that the top selling states are California, New York and Texas.

```python
## Profit vs Quantity
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df, color = 'Orange')
plt.legend()
plt.show()
```

```python
 
fig, axs = plt.subplots(ncols=2, nrows=2,figsize=(40,40))
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['yellow','red'],figsize=(15,10), ax = axs[0][0])
plt.ylabel('Profit/Loss and sales')

df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['yellow','green'],figsize=(15,10), ax = axs[0][1])
plt.ylabel('Profit/Loss and sales')

df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','Blue'],figsize=(15,10), ax = axs[1][0])
plt.ylabel('Profit/Loss and sales')

ps = df.groupby('Sub-Category')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['yellow','Black'],figsize=(15,10), ax = axs[1][1])
plt.ylabel('Profit/Loss and sales')

axs[0][0].set_title('Segments', fontsize = 20  )
axs[0][1].set_title('Regions', fontsize = 20 )
axs[1][0].set_title('Category', fontsize = 20 )
axs[1][1].set_title('Sub-Category', fontsize = 20 )

plt.show()
```

## From the above Four graph we observe that
## Graph-1 : Profit and sales are maximum in the consumer segment and minimum in the home office segment.

## Graph-2 : Profit and Sales are maximum in the West Region and minimum in the South Region

## Graph-3 :Technology and Office Supplies have the highest profit wheras furniture has least profit.

## Graph-4 : 
## 1. Phones have high sales.

## 2. Chairs have less profit than sub-category phones.

## 3. Tables and Book-Cases are facing loss.
